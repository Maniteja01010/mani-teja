# Outliers

we have encontred some outliers in the above boxplot now we are treting them

```python
plt.figure(figsize=(15,10))
for features in ['Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Proanthocyanins', 'Color', 'Hue']:
  sns.boxplot(x=data[features])
  plt.tight_layout()
  plt.show()
```

```python
features = ['Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Proanthocyanins', 'Color', 'Hue']
for feature in features:
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[feature] >= lower_bound) & (data[feature] <= upper_bound)]
```

```python
data.head()
```

```python
plt.figure(figsize=(15,10))
for i,features in enumerate(features,1):
  plt.subplot(2,4,i)
  sns.boxplot(x=data[features])
  plt.title("After removing outliers")
plt.tight_layout()
plt.show()
```

Removeing Columns with All Missing Values

**Standardizeing the features**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(data_cleaned)
```

# Evaluate Clustering Performance

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA().fit(scaled_data)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1),np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.show()
```

W e got an optimal number of principal compantent using scree plot and the cumulaive explained variance plot

# PCA(Principal Component Analysis)

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=6)
pca_data=pca.fit_transform(scaled_data)
```

```python
pca_df=pd.DataFrame(pca_data,columns=['PC1','PC2','PC3','PC4','PC5','PC6'])
```

**This is Kmeans is on orignal data**

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3,random_state=42)
kmeans.fit(data_cleaned)
labels=kmeans.labels_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data_cleaned.iloc[:, 0], data_cleaned.iloc[:, 1], c=labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering on original data')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
```

```python
silhouette_avg=silhouette_score(data_cleaned,labels)
print(f'Silhouette Score: {silhouette_avg}')
```

```python
db_index = davies_bouldin_score(data_cleaned, labels)
print(f'Davies-Bouldin Index: {db_index}')
```

**This is Clustering with PCA Data**

```python
kmeans_pca=KMeans(n_clusters=3,random_state=42)
kmeans_pca.fit(pca_df)
pca_labels=kmeans_pca.labels_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(pca_df.iloc[:, 0],pca_df.iloc[:, 1],c=pca_labels,cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering on PCA Data')
plt.show()
```

```python
silhouette_avg_pca=silhouette_score(pca_df,pca_labels)
print(f'Silhouette Score: {silhouette_avg_pca}')
```

```python
db_index_pca = davies_bouldin_score(pca_df, pca_labels)
print(f'Davies-Bouldin Index: {db_index_pca}')
```

# Comparison and Analysis

```python
print(f'Silhouette Score (Original Data): {silhouette_avg}')
print(f'Davies-Bouldin Index (Original Data): {db_index}')
print(f'Silhouette Score (PCA Data): {silhouette_avg_pca}')
print(f'Davies-Bouldin Index (PCA Data): {db_index_pca}')
```

Comparison and Analysis
Clustering Results Comparison:
Original Data:

Silhouette Score: 0.4575 (indicating moderately well-defined clusters)

Davies-Bouldin Index: 0.713 (indicating fairly compact and well-separated clusters)

PCA-transformed Data:

Silhouette Score: 0.2561 (indicating less well-defined clusters)

Davies-Bouldin Index: 1.4055 (indicating less compact and more similar clusters)

Similarities and Differences:
Similarities:

Both datasets were clustered using the same K-means algorithm.

Clusters are identifiable in both cases.

Differences:

Clustering on the original data produced more distinct and better-defined clusters compared to the PCA-transformed data.

PCA-transformed data clustering resulted in higher overlap between clusters and less clear separation.

Impact of Dimensionality Reduction:
PCA:

Reduced dimensionality to simplify the dataset and make computations more efficient.

However, some vital information for clustering might have been lost, leading to less distinct clusters.

PCA helps in visualizing high-dimensional data but can sometimes affect clustering clarity.

Trade-offs:
Using PCA:

Pros: Simplifies data, reduces computational load, and aids in visualizing complex datasets.

Cons: Might lose some clustering-relevant information, leading to less distinct clusters.

Clustering on Original Data:

Pros: Retains all feature information, resulting in better-defined and more distinct clusters.

Cons: Computationally more expensive and harder to visualize if the data is high-dimensional.
Clustering on the original dataset provided better-defined clusters, as evidenced by higher Silhouette Scores and lower Davies-Bouldin Index values. While PCA simplifies data and reduces dimensionality, it can impact the clarity and quality of clustering

# Conclusion and Insights

**Key Findings:**

Exploratory Data Analysis (EDA):

Initial data exploration provided a clear understanding of feature distributions and correlations.

Dimensionality Reduction with PCA:

PCA effectively reduced the dataset’s dimensionality while retaining 95% of the variance.

The Scree Plot and Cumulative Explained Variance Plot guided the selection of 6 principal components.

Clustering with Original Data:

K-means clustering on the original dataset resulted in well-defined clusters with a Silhouette Score of 0.4575 and a Davies-Bouldin Index of 0.713.

Clustering with PCA Data:

Clustering on the PCA-transformed data was less distinct, with a Silhouette Score of 0.2561 and a Davies-Bouldin Index of 1.4055.

**Practical Implications:**

Dimensionality Reduction:

PCA simplifies high-dimensional datasets, making computations more efficient and aiding in visualizing data.

However, there is a trade-off as some information critical for clustering can be lost, affecting the quality of the clusters.

Clustering:

Clustering directly on the original dataset retains all feature information, producing more distinct and better-defined clusters.

When computational resources or visualization are a priority, PCA followed by clustering can still provide meaningful insights, despite some loss in clustering quality.

# Recommendations:

When to Use PCA:

Use PCA when dealing with very high-dimensional data where computational efficiency and visualization are important.

Suitable for exploratory data analysis to identify underlying patterns and reduce noise.

When to Cluster on Original Data:

Opt for clustering on the original dataset when the goal is to achieve the most distinct and well-defined clusters.

Best for cases where the full feature set is crucial for accurate clustering and the data’s dimensionality is manageable.
